```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import cv2
```

Menambahkan library matplotlib.pyplot untuk digunakan dalam visualisasi data

Menambahkan library numpy untuk proses matematika yang dilakukan dalam pengolahan citra

Menambahkan library pandas untuk menganalisis dan memanipulasi data yang berkaitan dengan citra

Menambahkan library cv2 yang berfungsi untuk pengolahan sebuah citra

Kemudian menginisialisasi image = plt.imread('aang.jpg') untuk menyiapkan sebuah citra aang.jpg kedalam variabel gambar untuk mempermudah pengolahan citra. Kemudian menambahkan print(image) untuk menampilkan nilai array yang terdapat pada variabel gambar dan plt.imshow(image) untuk menampilkan tampilan citra dari varibel gambar

```python
image = plt.imread('aang.jpg')
print(image)
plt.imshow(image)
```

```python
blue_channel, green_channel, red_channel = cv2.split(image)
print("Nilai piksel dari gambar asli (kanal biru):")
print(blue_channel)

print("\nNilai piksel dari gambar asli (kanal hijau):")
print(green_channel)

print("\nNilai piksel dari gambar asli (kanal merah):")
print(red_channel)
```

Code di atas merupakan percobaan untuk mengetahui citra RGB dari gambar.
blue_channel, green_channel, red_channel = cv2.split(image) digunakan untuk memisahkan citra yang disimpan dalam variabel image menjadi tiga kanal warna (merah, hijau dan biru), kemudian melakukan print untuk menvetak nilai piksel dalam setiap kanal yang nilainya direpresentasikan dalam bentuk array Numpy

Kemudian setelah menampilkan gambar maka selanjutnya adalah mengubah gambar menjadi Grayscale dengan menggunakan
aangAbu = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY) untuk membuat variabel baru bernama "aangAbu" untuk menyimpan citra pada variabel gambar yang telah diolah dimensi warnanya dari RGB ke GRAY

```python
#Grayscale
aangAbu = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
print("Nilai piksel gambar grayscale: ")
print(aangAbu)
plt.imshow(aangAbu,cmap='gray')
```

Kemudian melakukan normalisasi, yaitu dengan pertama menghitung nilai max dan min gambar menggunakan fungdi np.max() dan np.min() untuk memberikan rentang nilai piksel pada gambar.  Kemudian melakukan normalisasi dengan rumus normalisasi, kemudian menampilkan hasilnya dengan plt.imshow()

```python
#Normalisasi
max_val = np.max(aangAbu)
min_val = np.min(aangAbu)

normalisasi = np.round((aangAbu - min_val) / (max_val - min_val)*255).astype(int)
plt.title("Citra Hasil Normalisasi")
print("Nilai piksel setelah normalisasi: ")
print(normalisasi)
plt.imshow(normalisasi, cmap='gray')
```

```python
# Jumlah intensitas Setiap Indeks Gambar
tinggi, lebar = normalisasi.shape
normalisasi.shape

sumGray = np.zeros(256, dtype=int)
for x in range(tinggi):
    for y in range(lebar):
        nilai_piksel = normalisasi[x, y]
        sumGray[nilai_piksel] += 1

grayvalue = [i for i in range(256)] 

print("Dimensi citra grayscale:", tinggi, lebar)
print("Jumlah intensitas setiap indeks gambar:")
for i in range(len(sumGray)):
    print("Histogram [", i, "]", sumGray[i])
```

Code di atas berguna untuk mengetahui jumlah intensitas setiap indeks gambar dengan menggunakan normalisasi.shape untuk mendapatkan tinggi dan lebar gambar. Kemudian sumGray sebagai array dengan panjang 256 (intensitas 0-255) dan grayvalue sebagai daftar nilai intensitas dari 0-255.

Kemudian dilakukan iterasi dimana setiap nilai piksel diambil dari gambar yang telah dinormalisasi kemudian digunakan sebagai indeks untuk menambahkan jumlah intensitas pada indeks tersebut di dalam array sumGray

Setelah perhitungan selesai, hasilnya dicetak. Setiap indeks histogram (nilai intensitas dari 0 hingga 255) beserta jumlah intensitasnya ditampilkan sebagai output.

```python
#Jumlah Intensitas Gambar
sumofIntensity = np.sum(sumGray)
print(sumofIntensity)
```

Code di atas untuk mengetahui jumlah intensitas gambar dengan np.sum() untuk menjumlahkan semua nilai yang ada di dalam sumGray

Kemudian menampilkan histogram berdasarkan frekuensi dan intensitas cintra gambar

```python
plt.bar(grayvalue, sumGray, color='green')
plt.xlabel('grayvalue (intensitas)')
plt.ylabel('sumGray (frekuensi)')
```

Kemudian memunculkan histogram berdasarkan intensitas dan fekuensi yang sudah dinormalisasi

```python
plt.bar(grayvalue, sumGray/(tinggi*lebar), color='green')
plt.xlabel('grayvalue (intensitas)')
plt.ylabel('sumGray (frekuensi normalisasi)')
```

SOAL NOMOR 2 (Konvolusi)

```python
kernel = [[1, 2, -1, 1],
          [0, 2, 1, 2],
          [2, 1, 1, 0],
          [1, 0, -1, 0]]
```

membuat variabel kernel untuk digunakan dalam konvolusi

Kemudian membuat zero padding pada image grayscale dengan menggunakan image_pad = np.pad(aangAbu, [(0,3), (0,3)]).astype(int), dimana zero padding yang dibuat adalah 3 baris di bawah dan 3 kolom di kanan


```python
#Zero Padding
#image_pad = np.pad(aangAbu, [(1, 1), (1, 1)], mode='constant', constant_values=0)
image_pad = np.pad(aangAbu, [(0,3), (0,3)]).astype(int)
# Tampilkan citra dengan zero padding
print(image_pad)
plt.title('Image with Zero Padding (semua)')
plt.imshow(image_pad, cmap='gray')
```

Kemudian melakukan konvolusi dari gambar grayscale zero padding dengan menggunakan rumus konvolusi yang di iterasi


```python
#konvolusi
konvolusi_aang = np.zeros(aangAbu.shape)

for x in range(tinggi):
    for y in range(lebar):
        konvolusi_aang[x][y] = kernel[0][0]*image_pad[x][y] + kernel[1][0]*image_pad[x+1][y] + kernel[2][0]*image_pad[x+2][y] + kernel[3][0]*image_pad[x+3][y] + kernel[0][1]*image_pad[x][y+1] + kernel[1][1]*image_pad[x+1][y+1] + kernel[2][1]*image_pad[x+2][y+1] + kernel[3][1]*image_pad[x+3][y+1] + kernel[0][2]*image_pad[x][y+2] + kernel[1][2]*image_pad[x+1][y+2] + kernel[2][2]*image_pad[x+2][y+2] + kernel[3][2]*image_pad[x+3][y+2] + kernel[0][3]*image_pad[x][y+3] + kernel[1][3]*image_pad[x+1][y+3] + kernel[2][3]*image_pad[x+2][y+3] + kernel[3][3]*image_pad[x+3][y+3]

plt.title("Hasil Konvolusi dengan Zero Padding")
print(konvolusi_aang)
plt.imshow(konvolusi_aang, cmap='gray')
```

Kemudian melakukan normalisasi terhadap pixel gambar grayscale zero padding yang telah dikonvolusi

```python
#Normalisasi
max_val = np.max(konvolusi_aang)
min_val = np.min(konvolusi_aang)

normalisasi_zp = np.round((konvolusi_aang - min_val)/(max_val - min_val))
plt.title("normalisasi konvolusi dengan zero padding")
print(normalisasi_zp)
plt.imshow(normalisasi_zp, cmap='gray')
```

Kemudian melakukan percobaan menampilkan gambar hasil konvolusi dan gambar konvolusi yang telah di normalisasi, gambar ditampilkan dengan subplot yang bisa menampilkan gambar lebih dari satu

```python
plt.figure(figsize=(10, 8))  

plt.subplot(1, 2, 1) 
plt.imshow(konvolusi_aang, cmap='gray')
plt.title('konvolusi')
plt.axis('off')

plt.subplot(1, 2, 2) 
plt.imshow(normalisasi_zp, cmap='gray')
plt.title('Normalisasi')
plt.axis('off')
```

SOAL NOMOR 3 (Translasi dan Rotasi)

Kemudian melakukan rotasi dengan sudut 40 derajat, dimana pertama kali dilakukan adalah membuat matriks kosong (imgRotate) yang akan menampung hasil rotasi gambar yang memiliki ukuran yang sama dengan gambar awal (aangAbu)

Kemudian    dibuat nilai tengah gambar sebagai pusat rotasi yang akan dihitung sebagai midRow dan midCol untuk menghitung koordinat piksel setelah dirotasi

kemudian terdapat iterasi untuk merotasi gambar dengan rumus rotasi, setelah itu dilakukan pengecekan koordinat setelah mendapat koordinat baru apakah berada dalam batas gambar yang baru sesuai dengan ukuran gambar hasil rotasi. Jika koordinat (x, y) berada dalam batas gambar, maka nilai piksel pada koordinat tersebut diambil dari gambar awal aangAbu dan disalin ke gambar hasil rotasi imgRotate.

```python
#Rotasi dengan sudut 40 derajat
imgRotate = np.zeros((tinggi, lebar)).astype(int)
sudut = np.deg2rad(40)
midRow = tinggi/2
midCol = lebar/2

for i in range (tinggi):
    for j in range(lebar):
        #x = int((x-centerX)*np.cos(sudut) - (y-centerY)*np.sin(sudut) + centerX)
        #y = int((x-centerX)*np.sin(sudut) + (y-centerY)*np.sin(sudut) + centerY)
        y = int((i - midRow)*np.cos(sudut)-(j-midCol)*np.sin(sudut) + midRow)
        x = int((i - midRow)*np.sin(sudut)+(j-midCol)*np.cos(sudut) + midCol)

        #y = int(y + midCol)
        #x = int(x + midRow)

        if x>=0 and x<lebar and  y>=0 and y<tinggi:
        #0<=x and x<lebar and 0<=y and y<tinggi:
            imgRotate[i][j] = aangAbu[y][x]


plt.imshow(imgRotate, cmap='gray')
```

Kemudian melakukan translasi, langkah awalnya hampir sama dengan proses yang lain yaitu membuat matriks kosong untuk menampung hasil translasi, kemudian dilakukan looping ( i = baris dan j = kolom dari gambar)
Kemudian dilakukan pengecekan apakah koordinat setelah translasi berada dalam batas gambar yang baru Jika koordinat tersebut berada dalam batas gambar yang baru, maka nilai piksel pada koordinat tersebut diambil dari gambar hasil rotasi (imgRotate) dan disalin ke gambar hasil translasi (imgTranslasi)

Dilakukan juga pengecekan agar nilai indeks tidak melebihi batas gambar yang baru


```python
#translasi sejauh x = 300 dan y = -163 
x = 300
y = -163
imgTranslasi = np.zeros((tinggi, lebar)).astype(int) 

for i in range(tinggi):
    for j in range(lebar):
        if (i-300>0 and j+163<lebar):
            imgTranslasi[i,j]=imgRotate[i-300,j+163]


plt.imshow(imgTranslasi, cmap="gray")
```

SOAL NOMOR 4 (Pencerminan dan Dilatasi)

Selanjutnya adalah melakukan pencerminan dengan menggunakan iterasi dan pencerminan dimana nilai piksel dari gambar imgTranslasi dipindahkan ke gambar imgMirrored dengan cara melakukan pencerminan ke arah sumbu Y. Proses pencerminan dilakukan dengan menyalin nilai piksel dari imgTranslasi dengan indeks kolom yang dibalik. Misalnya, jika indeks kolom di imgTranslasi adalah x, maka indeks kolom yang sesuai di imgMirrored adalah lebar - x - 1, dimana lebar adalah lebar gambar. Ini menyebabkan gambar terpantulkan secara horizontal.

```python
#Pencerminan ke Arah sumbu Y

imgMirrored = np.zeros((tinggi, lebar)).astype(int)

for y in range(tinggi):
	for x in range(lebar):
		imgMirrored[y,x] = imgTranslasi[y,lebar-x-1]

plt.imshow(imgMirrored, cmap='gray')
```

Kemudian melakukan dilatasi dengan skala 3, dimana dengan membuat matrik kosong yang berukurang 3 kali lebih besar dahulu, kemudian melakukan iterasi untuk setiap piksel dan melakukan dilatasi dimana nilai piksel dari gambar yang telah di mirroring dipindahkan ke matriks kosong yang sudah dibuat sebanyak tiga kali baik dalam arah vertikal maupun horizontal, sehingga menghasilkan gambar yang terdilatasi.

```python
#Dilatasi dengan skala 3
imgDilatasi = np.zeros((tinggi*3, lebar*3)).astype(int)

for i in range(tinggi*3):
	for j in range(lebar*3):
		imgDilatasi[i,j] = imgMirrored[int(i/3),int(j/3)]

plt.imshow(imgDilatasi, cmap='gray')
```
